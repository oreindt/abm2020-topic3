Two key attributes:
- whether two individuals have the same or opposite belief
- whether two individuals see the information from the same angle

# TWO MAJOR RULES FOR UPDATING BELIEFS

If two people have the same belief & if they see things from the same angle
They will experience self-reinforcing
        
        `if (sign(person.belief) == sign(other.belief) || mid_belief(other)) &  (person.info == other.info)
            other.belief = (person.belief + other.belief) / 2`


If two people have opposite belief & if they see things from the same angle
They will have an exchange in view and get their view closer to the truth
        
        `if (sign(person.belief) != sign(other.belief) || mid_belief(other)) &  (person.info == other.info)
            other.belief = other.belief - abs(person.belief)`

Now, we see a different updating rule

If two people have opposite belief & if they see things from the same angle
They will have an exchange in view and get their view closer to the truth
        
        `if (sign(person.belief) != sign(other.belief) || mid_belief(other)) &  (person.info == other.info)
            other.belief = other.belief - abs(person.belief)/100`
    
Note that we have `- abs(person.belief)/100`, a much smaller "breaking myth" effect (divided by 100).
