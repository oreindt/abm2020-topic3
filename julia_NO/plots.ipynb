# Plots

Use the `Plots` package.

```julia
using Plots; gr() # Using the Plotly Backend
```

Include the file with the simulation setup.

```julia
include("setup.jl")
```

Create a plot.

```julia
sim = setup_sim(p=0.1, N=1000, p_c=1.0, n_inf=10, b=0, seed=42)
knowing, pos, neg, mid, avg_belief, abs_avgs = run_sim(sim, 500)

Plots.plot([knowing pos neg mid avg_belief abs_avgs], labels = ["knowing" "highly pos" "highly neg" "mid" "avg belief" "abs average"])
```

```julia
Plots.plot([pos neg mid], labels = ["highly pos" "highly neg" "mid"])
```
